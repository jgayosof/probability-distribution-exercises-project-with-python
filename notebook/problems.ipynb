# Probability Distribution Problems

**Binomial Distribution**

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either Head or Tails.

```python
# Probability functions, density and distribution

def dbinom(x,size,prob=0.5):
    """
    Calculates the point estimate of the binomial distribution -- PMF
    """
    from scipy.stats import binom
    result=binom.pmf(k=x,n=size,p=prob,loc=0)
    return result

def pbinom(q,size,prob=0.5):
    """
    Calculates the cumulative of the binomial distribution -- CDF
    """
    from scipy.stats import binom
    result=binom.cdf(k=q,n=size,p=prob,loc=0)
    return result

def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution --PPF?
    """
    from scipy.stats import binom
    result=binom.ppf(q=p,n=size,p=prob,loc=0)
    return result

def rbinom(n,size,prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.rvs(n=size,p=prob,size=n)
    return result
```

```python
dbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) takes the value 2,
pbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) will take a value less than 2
qbinom(0.9, size = 10, prob = 0.2) # what value of a binomial (10,0.2) has a cumulative probability of 0.9?
rbinom(2000, size = 10, prob = 0.2) # generate 2000 random values from a binomial distribution (10,0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete distributions
xs = np.arange( 1, 11 )
ys = dbinom(xs,size=10,prob=0.2) # compute the shape of the distribution
plt.plot( xs, ys, 'o' )   # plot circles...
plt.vlines( xs, 0, ys )   # ...and lines
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function binomial distribution')
plt.show()
```

```python
#Code here

# NORMAL DISTRIBUTION (mean, std)
def normal_pdf(x, mean=0, std=1) :
    """
    Calculates the point estimate of the normal distribution -- PDF
    """
    from scipy.stats import norm
    result = norm.pdf(x, loc=mean, scale=std)
    return result

def normal_cdf(x, mean=0, std=1) :
    """
    Calculates the cumulative of the normal distribution -- CDF
    """
    from scipy.stats import norm
    result = norm.cdf(x, loc=mean, scale=std)
    return result

def normal_quantile(prob, mean=0, std=1) :
    """
    Calculates the quantile function from the normal distribution
    x / p(X<=x) = prob
    """
    from scipy.stats import norm
    result = norm.ppf(q=prob, loc=mean, scale=std)
    return result

def normal_random_numbers(n, mean=0, std=1):
    """
    Generates random variables from the normal distribution
    """
    from scipy.stats import norm
    result = norm.rvs(loc=mean, scale=std, size=n)
    return result

# Test Prints:
x = 2
prob = 0.6
n = 5
print('NORMAL')
print('normal_pdf for x = ' +str(x)+ ' is ' +str(normal_pdf(x=x, mean=0, std=1)))
print('normal_cdf for x = ' +str(x)+ ' is ' +str(normal_cdf(x=x, mean=0, std=1)))
print('normal_quantile for prob = ' +str(prob)+ ' is ' +str(normal_quantile(prob=prob, mean=0, std=1)))
print(str(n)+ ' normal random numbers: '  +str(normal_random_numbers(n=n, mean=0, std=1)))
print('\n')


# UNIFORM DISTRIBUTION (mean, std)
def uniform_pdf(x, a=0, b=1) :
    """
    Calculates the point estimate of the uniform distribution -- PDF
    """
    from scipy.stats import uniform
    result = uniform.pdf(x=x, loc=a, scale=(b-a))
    return result

def uniform_cdf(x, a=0, b=1) :
    """
    Calculates the cumulative of the uniform distribution -- CDF
    """
    from scipy.stats import uniform
    result = uniform.cdf(x, loc=a, scale=(b-a))
    return result

def uniform_quantile(prob, a=0, b=1) :
    """
    Calculates the quantile function from the uniform distribution
    """
    from scipy.stats import uniform
    result = uniform.ppf(q=prob, loc=a, scale=(b-a))
    return result

def uniform_random_numbers(n, a=0, b=1) :
    """
    Generates random variables from the uniform distribution
    """
    from scipy.stats import uniform
    result = uniform.rvs(size=n, loc=a, scale=b)
    return result

# Test Prints:
a = 1
b = 2
x = -1,0.2,0.8,2
prob = 0.6
n = 5
print('UNIFORM')
print('uniform_pdf for x = ' +str(x)+ ' is ' +str(uniform_pdf(x=x, a=a, b=b)))
print('uniform_cdf for x = ' +str(x)+ ' is ' +str(uniform_cdf(x=x, a=a, b=b)))
print('uniform_quantile for prob = ' +str(prob)+ ' is ' +str(uniform_quantile(prob=prob, a=a, b=b)))
print(str(n)+ ' uniform random numbers: '  +str(uniform_random_numbers(n=n, a=a, b=b)))
```
